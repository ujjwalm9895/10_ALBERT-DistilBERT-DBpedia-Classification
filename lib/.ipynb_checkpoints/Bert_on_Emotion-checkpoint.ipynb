### Import Libraries:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import ktrain
from ktrain import text
import tensorflow as tf
from sklearn.model_selection import train_test_split
from datasets import list_datasets
from datasets import load_dataset
import timeit
import warnings

pd.set_option('display.max_columns', None)
warnings.simplefilter(action="ignore")
```

```python
tf.__version__
```

### Load Dataset:

```python
emotion_train = load_dataset('emotion', split='train')
emotion_val = load_dataset('emotion', split='validation')
```

```python
print("\nTrain Dataset Features for Emotion: \n", emotion_train.features)
print("\nValidation Dataset Features for Emotion: \n", emotion_val.features)
```

```python
emotion_train_df = pd.DataFrame(data=emotion_train)
emotion_val_df = pd.DataFrame(data=emotion_val)
```

```python
class_label_names = ['sadness', 'joy', 'love', 'anger', 'fear', 'surprise']
```

### Split Train & Validation data:

```python
X_train = emotion_train_df[:]["text"]
y_train = emotion_train_df[:]["label"]
X_test = emotion_val_df[:]["text"]
y_test = emotion_val_df[:]["label"]

print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

### Instantiating a BERT Instance:
Create a BERT instance with the model name, max token length, the labels to be used for each category and the batch size.

```python
bert_transformer = text.Transformer('bert-base-uncased', maxlen=512, classes=class_label_names, batch_size=6)
```

### Perform Data Preprocessing:

```python
bert_train = bert_transformer.preprocess_train(X_train.to_list(), y_train.to_list())
bert_val = bert_transformer.preprocess_test(X_test.to_list(), y_test.to_list())
```

### Compile BERT in a K-Train Learner Object:
Since we are using k-train as a high level abstration package, we need to wrap our model in a k-train Learner Object for further compuation

```python
bert_model = bert_transformer.get_classifier()
```

```python
bert_learner_ins = ktrain.get_learner(model=bert_model,
                            train_data=bert_train,
                            val_data=bert_val,
                            batch_size=6)
```

### BERT Model Summary:

```python
bert_learner_ins.model.summary()
```

### BERT Optimal Learning Rates:

As per the research paper on BERT, below are the details for optimal learning rates:

* Batch Sizes => {16, 32}
* Learning Rates => {1e−5, 2e−5, 3e−5}

### Fine Tuning BERT on Emotion Dataset:
We take our emotion dataset along with the BERT model we created, define the learning-rate & epochs to be used and start fine-tuning.

```python
bert_fine_tuning_start= timeit.default_timer()
bert_learner_ins.fit_onecycle(lr=2e-5, epochs=3)
bert_fine_tuning_stop = timeit.default_timer()
```

```python
print("\nFine-Tuning time for BERT on Emotion dataset: \n", (bert_fine_tuning_stop - bert_fine_tuning_start)/60, " min")
```

### Checking BERT performance metrics:

```python
bert_validation_start= timeit.default_timer()
bert_learner_ins.validate()
bert_validation_stop= timeit.default_timer()
```

```python
print("\nInference time for BERT on Emotion dataset: \n", (bert_validation_stop - bert_validation_start), " sec")
```

```python
bert_learner_ins.validate(class_names=class_label_names)
```

```python
bert_learner_ins.view_top_losses(preproc=bert_transformer)
```

### Saving BERT Model:

```python
bert_predictor = ktrain.get_predictor(bert_learner_ins.model, preproc=bert_transformer)
bert_predictor.get_classes()
```
